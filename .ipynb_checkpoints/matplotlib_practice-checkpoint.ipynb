```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)

# Can also be written like this:
    
# fig = plt.figure()
# ax = fig.add_subplot()
```

```python
#Can also be written like this:
    
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

ax = plt.axes()

ax.plot(x_axis, y_axis)

#To change figure-level attributes (such as axis labels, a title, or a legend) or save the figure as an image, use fig = plt.figure()
```

```python
plt.plot(x_axis, y_axis)

#only use plt.show() once per session, usually at end
```

```python
# Create the plot and add a label for the legend, plus colors , marker points, and width of lines.

plt.plot(x_axis, y_axis,  marker="*", color="blue", linewidth=2, label="Boston")

#create labels for x and y axis

plt.xlabel("Date")
plt.ylabel("Fare ($)")

#set y limit between 0 and 45

plt.ylim(0, 45)

#set title

plt.title("PyBer Fare by Month")

#add a grid

plt.grid()

#add a legend

plt.legend()
```

```python
#challenge: make changes using object-oriented

# Create the plot and add a label for the legend, plus colors , marker points, and width of lines.

fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(x_axis, y_axis, label="Boston", color="green", marker="d", linewidth=2)

ax.set_title("PyBer Fare by Month")

ax.set_xlabel("Date")

ax.set_ylabel("Fare ($)")

ax.set_ylim(0, 45)

ax.grid()
            
```

```python
########### BUBBLE OOP #######

fig, ax = plt.subplots()

# ax.scatter(x_axis, y_axis, s=y_axis)

even_larger_y = [data*5 for data in y_axis]

ax.scatter(y_axis, x_axis, s=even_larger_y, color="skyblue", edgecolors="black", linewidths=2, label="Boston")

ax.legend()

ax.set_title("PyBer Fares by Month")

ax.invert_yaxis()

ax.set_xlim(0, 45)
```

```python
############## PIE CHARTS MATLAB ####################


colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

#explode is a tuple, length mush match length of labels list (wedges)

plt.subplots(figsize=(8, 8))

plt.pie(y_axis, 
        explode=explode_values, 
        labels=x_axis, 
        autopct="%.1f%%", 
        colors=colors)

#autopct is to format labels

# making chart bigger:

plt.show()
```

```python
############## PIE CHARTS OOP ####################

fig, ax = plt.subplots()

explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]


ax.pie(
       y_axis, 
       labels=x_axis,
       autopct="%.1f%%",
       explode=explode_values,
       colors=colors,
        shadow=True,
        startangle=95,
    counterclock=False,
    radius=2
)
plt.show()
```
